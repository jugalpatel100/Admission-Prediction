#### Seperating features and output

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

#### Feature scaling

```python
#Feature scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_x = scaler.fit_transform(x)
```

##### Applying PCA to reduce the no of dimensions to 4

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=4)

new_x = pca.fit_transform(scaled_x)
```

```python
explained_var = pd.DataFrame(list(enumerate(np.array(pca.explained_variance_ratio_)*100, start=1)), columns=['Principal components', 'explained var %'])
```

```python
explained_var
```

```python
fig, axes = plt.subplots(1,1,figsize=(6,5))
```

```python
fig.tight_layout(pad=8)
```

```python
sns.set_style('whitegrid')
```

```python
sns.barplot(x="Principal components", y="explained var %", data=explained_var, ax=axes, color='#199ae0')
axes.set_xlabel("Principal components", color="#838a0b", fontsize=8)
axes.set_ylabel("Percentage of explained variances", color="#838a0b", fontsize=8)
axes.set_title("Percenatge of variance for each Princpal Component", pad=20)
axes.set_yticks(list(range(0,110,10)))
```

#### Splitting the data

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(new_x, y, random_state=0)
```

### MLR with PCA

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()

lin_reg.fit(x_train, y_train)
```

```python
y_pred = lin_reg.predict(x_test)
```

```python
lin_reg.intercept_
```

```python
lin_reg.coef_
```

```python
print("Train score:", lin_reg.score(x_train, y_train))
print("Test score:", lin_reg.score(x_test, y_test))
```

```python
from sklearn.metrics import r2_score, mean_squared_error

print("R2-score                : %.2f" %r2_score(y_test, y_pred))
print("Mean squared error      : %.4f" %mean_squared_error(y_test, y_pred))
print("Root mean squared error : %.4f" %np.sqrt(mean_squared_error(y_test, y_pred)))
```

### RFR with PCA

```python
from sklearn.ensemble import RandomForestRegressor

reg = RandomForestRegressor(n_estimators=80, max_depth=4, min_samples_split=40, min_samples_leaf=10, random_state=100)
```

```python
reg.fit(x_train, y_train)
```

```python
feature_importance = pd.DataFrame(zip(["PC-"+str(i) for i in range(1,5)], reg.feature_importances_, ), columns=['Feature','Value'])
```

```python
feature_importance
```

```python
sns.set_style("whitegrid")
```

```python
fig, axes = plt.subplots(1,1, figsize=(10,5))
```

```python
fig.tight_layout(pad=5)
```

```python
sns.barplot(x='Value', y='Feature', data=feature_importance.sort_values(by='Value', ascending=False), orient='horizontal')
axes.set_xticks(np.arange(0,1.05,0.05))
axes.set_xlabel("Feature importance value", color="#838a0b", fontsize=8)
axes.set_ylabel("Principal components", color="#838a0b", fontsize=8)
axes.set_title("Feature importance of each PC", pad=20)
```
